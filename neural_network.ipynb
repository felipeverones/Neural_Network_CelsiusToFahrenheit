```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
layer = tf.keras.layers.Dense(units=1, input_shape=[1])  #Dense pra camada fully connected

#criando o modelo
model = tf.keras.Sequential([layer])
```

```python
model.compile(
    optimizer= tf.keras.optimizers.Adam(0.1), #algoritmo Adam, responsável por ir melhorando aos poucos; 0.1 é quanto o algoritmo ajusta os pesos
    loss = 'mean_squared_error' #loss function
)
```

```python
results = model.fit(celsius, fahrenheit, epochs=100000, verbose=False)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("nº epochs")
plt.ylabel("Loss")
plt.plot(results.history["loss"])
```

```python
myinput = np.array([100.0]).reshape(-1, 1)  # Reshape para garantir 2 dimensões, onde a primeira é o lote

res = model.predict(myinput)

print("O resultado é " + str(res) + "°F (Fahrenheit)")
```
